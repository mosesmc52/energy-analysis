### Example of a Ornstein-Uhlenbeck mean reverting process 

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters for the Ornstein-Uhlenbeck process
mu = 0.0      # Long-term mean
theta = 0.15  # Speed of reversion
sigma = 0.3   # Volatility (standard deviation of random fluctuations)
dt = 0.01     # Time step
T = 1.0       # Total time
x0 = 1.0      # Initial value

# Time grid
N = int(T / dt)
t = np.linspace(0, T, N)

# Simulate the OU process
x = np.zeros(N)
x[0] = x0
for i in range(1, N):
    x[i] = x[i-1] + theta * (mu - x[i-1]) * dt + sigma * np.sqrt(dt) * np.random.normal()

```

```python
# Plot the results
plt.plot(t, x)
plt.title('Ornstein-Uhlenbeck Process')
plt.xlabel('Time')
plt.ylabel('X(t)')
plt.grid(True)
plt.show()
```
