```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define U.S. recession periods
recession_periods = [('2001-03', '2001-11'),
                     ('2007-12', '2009-06'),
                     ('2020-02', '2020-04')]

# Create example time series data (e.g., GDP or unemployment rate)
np.random.seed(0)
dates = pd.date_range('1990-01-01', periods=365, freq='M')  # Monthly data from 1990 to 2025
data = np.random.randn(len(dates)).cumsum()  # Cumulative random data to simulate a time series



# Plot the time series
plt.plot(dates, data, label='Economic Indicator')

# Highlight recession periods
for start, end in recession_periods:
    plt.axvspan(pd.to_datetime(start), pd.to_datetime(end), color='red', alpha=0.3, label='Recession Period')

# Ensure no duplicate labels for the shaded areas
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Economic Indicator with U.S. Recession Periods Highlighted')
plt.grid(True)

# Show the plot
plt.show()
```
