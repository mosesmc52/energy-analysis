```python
# Define periods of high inflation (e.g., above a certain threshold, say 6%)
high_inflation_periods = [('2008-01', '2009-01'), 
                          ('2021-01', '2022-01')]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create example time series data (e.g., inflation rates)
np.random.seed(0)
dates = pd.date_range('2000-01-01', periods=240, freq='M')  # Monthly data for 20 years
inflation_rate = np.random.rand(len(dates)) * 10  # Random inflation rates between 0 and 10%


# Plot the inflation rate time series
plt.plot(dates, inflation_rate, label='Inflation Rate (%)', color='blue')

# Highlight periods of high inflation
for start, end in high_inflation_periods:
    plt.axvspan(pd.to_datetime(start), pd.to_datetime(end), color='orange', alpha=0.5, label='High Inflation Period')

# Ensure no duplicate labels for the shaded areas
handles, labels = plt.gca().get_legend_handles_labels()
by_label = dict(zip(labels, handles))
plt.legend(by_label.values(), by_label.keys())

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Inflation Rate (%)')
plt.title('Inflation Rate with High Inflation Periods Highlighted')
plt.grid(True)

# Show the plot
plt.show()
```
